## ASR Usage Example

```python
# Import the ASRProcessor class
from asr.asr import ASRProcessor

# Create an instance of ASR processor (small whisper model)
asr = ASRProcessor(model_name= "small")

# Set the path of the test audio question
audio_input = "asr/test_question.m4a"

# Get and print the result
result = asr.encode_audio(audio_input)
print(f"Transribed Text: {result['text']}")
```

## TTS Usage Example

```python
# Import the TTSProcessor class
from tts.tts import TTSProcessor

# Import sound files to hear the output
import soundfile as sf
from IPython.display import Audio

# Create an instance of the TTS processor
tts = TTSProcessor()

# Set example text (to be replaced with output from the assistant)
text = "If you are hearing this spoken aloud, then text-to-speech is working properly!"

# Save and play the response
waveform = tts.decode_audio(text)
sf.write("output_random_voice.wav", waveform.cpu().numpy(), 16000)
Audio("output_random_voice.wav")
```
